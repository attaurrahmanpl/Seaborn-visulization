```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
data_frame.head()
```

```python
data_frame.describe()
```

# Creating hoistogram

```python
data_frame.isnull().sum()
```

```python
# in python we call displot to histogram
y = data_frame['tip']
sns.distplot(y)
plt.show()
```

```python
sns.distplot(data_frame.total_bill)
plt.show()
```

```python
sns.distplot(data_frame['tip'] , vertical=False , color="red" , rug_kws = {"color": "g"}, bins =10)
#sns.distplot(data['sex'])
plt.show()
```

```python
sns.distplot(data_frame['total_bill'] , hist = False)
plt.show()
```

```python
# kde Whether to plot a gaussian kernel density estimate.
sns.distplot(data_frame['total_bill'] , kde = True , rug=False )
plt.show()
```

```python
sns.distplot(data_frame.total_bill , kde = False)
plt.show()
```

```python
sns.distplot(data_frame['total_bill'] , hist = True , color = 'r' , norm_hist = False , axlabel = 'Total bill'  , label = "total_bill")
plt.legend()
plt.title('Histogram' , color = 'r' , fontweight='bold' , fontsize = "20")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set(style='dark')
sns.distplot(data_frame['total_bill'])
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set(style='dark')
sns.distplot(data_frame['total_bill'] , axlabel = 'Total bill',
            hist_kws={'color':'#1A012D','edgecolor':'#BA2322' , 'linestyle':'--' , 'alpha':0.9 })
plt.show()
```

```python
plt.figure(figsize=(16 , 9))
sns.set(style = 'dark')
sns.distplot(data_frame['total_bill'] , axlabel='Total bill',
            hist_kws = {'color' : '#1A012D' , 'edgecolor' : '#BA2322' , 'linestyle' : '--' , 'alpha':0.9},
            kde_kws = {'color' : '#53A6DC' , 'linestyle' : '--' , 'alpha':0.9},
            rug = True,
            rug_kws = {'color' : '#53A6DC' , 'linestyle' : '--' , 'alpha':0.9})
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set(style='dark')
sns.distplot(data_frame['total_bill'] , axlabel='Total bill' ,
            hist_kws = {'color':'#1A012D' , 'edgecolor':'#BA2322' ,'linestyle' : '--' , 'alpha' : 0.9},
            fit_kws = {'color' : 'ffffff' , 'edgecolor' : '#BA2322' , 'linestyle' : '--' , 'alpha' : 0.1},
            rug=True,
            rug_kws = {'color' : '#53A6DC' , 'linestyle' : '--' , 'alpha':0.9})
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set(style = 'dark')
sns.distplot(data_frame['total_bill'] , axlabel = 'Total bill',
            hist_kws = {'color' : '#1A012D' , 'edgecolor' : '#BA2322' , 'linestyle' : '--' , 'alpha' : 0.9},
            fit_kws = {'color' : '#1A012D' , 'edgecolor' : '#BA2322' , 'Linewidth' : 5 , 'linestyle' : '--' , 'alpha' : 0.9},
            rug = False,
            rug_kws = {'color' : '#53A6DC' ,  'linestyle' : '--' , 'alpha' : 0.9})
plt.show()
```
