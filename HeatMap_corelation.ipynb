```python
df = pd.read_csv('https://raw.githubusercontent.com/attaurrahmanpl/all-csv/main/Who_is_responsible_for_global_warming.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
#  to drop the feature having char categorical data
#  drop the categorical feature and get country name as index
numer_df = df.drop(columns=['Country Code' , 'Indicator Name' , 'Indicator Code']).set_index('Country Name')
numer_df.head()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(numer_df)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(numer_df.head(20),vmin = 0, vmax = 21,cmap='Greens')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(numer_df.head(20),vmin = 0, vmax = 21,cmap='Greens', annot = True)
plt.show()
```

```python
annot_kws = {'fontsize':8, 
           'fontstyle':'italic',  
           'color':"r",
           'alpha':0.9, 
           'rotation':"horizontal",
           'verticalalignment':'center',
           'backgroundcolor':'w'}
plt.figure(figsize=(16,9))
sns.heatmap(numer_df.head(20),vmin = 0, vmax = 21, annot_kws=annot_kws, annot=True)
plt.show()
```

```python
#  to draw lines in on the heatmap
annot_kws = {'fontsize':8, 
           'fontstyle':'italic',  
           'alpha':0.9, 
           'rotation':"horizontal",
           'verticalalignment':'center'}
plt.figure(figsize=(16,9))
sns.heatmap(numer_df.head(20),vmin = 0, vmax = 21,annot_kws=annot_kws,annot=True,linewidths=1)
plt.show()
```

```python
#  to remove x labels y labals and bar from the map
plt.figure(figsize=(16,9))
sns.heatmap(numer_df.head(20),vmin = 0, vmax = 21,annot=True,cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```

```python
# desing a color bar
cbar_kws = {"orientation":"vertical", 
            "shrink":1,
            'extend':'max', 
            'extendfrac':0.2, 
            "ticks":np.arange(0,22), 
            "drawedges":True,
           }
#Drop specific record
# data22 = numer_df.drop(['American Samoa'])
plt.figure(figsize=(16,9))
# plt.title('Who is responsible for global warming'),
# plt.xlabel('CO2 emission per year')
# plt.ylabel('Country Name')
heatmap = sns.heatmap(numer_df.head(20),vmin = 0, vmax = 21,annot=True,cbar_kws=cbar_kws)

heatmap.set(title=('Who is responsible for global warming'),
           xlabel=('CO2 emission per year'),
           ylabel=('Country name'))

plt.show()
```

```python
# correlation
# sns heatmap correlation
 
plt.figure(figsize=(16,9))
 
ax = sns.heatmap(numer_df.corr(), annot = True,linewidths=1)
# ax.tick_params(direction='out', length=9, width=1, colors='g',
#                grid_color='b', grid_alpha=0.9,labelsize=10,labelcolor='red')
plt.title("Who is responsible for global warming")
plt.xlabel("CO2 emission per year")
plt.ylabel("country")
plt.show()
```

```python
# correlation
# sns heatmap correlation
# play with tick perameters
plt.figure(figsize=(16,9))
 
ax = sns.heatmap(numer_df.corr(), annot = True,linewidths=1)
ax.tick_params(direction='inout', length=10, width=2, colors='b',
               grid_color='b', grid_alpha=0.9,labelsize=10,labelcolor='red')
plt.title("Who is responsible for global warming")
plt.xlabel("CO2 emission per year")
plt.ylabel("country")
plt.show()
```
