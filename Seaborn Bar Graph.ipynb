```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
x = data['sex']
y = data['tip']
sns.barplot(x , y , data = data)
plt.show()
```

```python
plt.figure(figsize=(10,5))
# x = data['day']
# y = data['total_bill']
sns.barplot(data['day'] , data['total_bill'] , data=data)
plt.show()
```

```python
plt.figure(figsize=(16,9))
#plt.legend('day')
# sns.barplot(x='day', y = 'total_bill', data = data , hue=data.sex)
sns.barplot(data['day'] , data['total_bill'] , hue = data['sex'])
# plt.legend(data.day)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.legend('day')
sns.barplot(y='day' , x= 'size' , data = data , hue=data.sex ,
            estimator=np.max , )
plt.show()
```

```python
sns.set()
kwargs = {'alpha':0.8,'linestyle':'--','linewidth':1.9,'edgecolor':'b'}
plt.title("bar plot")
sns.barplot(x='day', y = 'size', data = data , **kwargs)
plt.show()
```

```python
sns.barplot(x='day' , y = 'total_bill' , data = data , hue=data.sex , alpha = 0.99)
plt.title("bar plot of tips dataset")
plt.xlabel('Day')
plt.ylabel('Totl bill')
plt.show()
```
