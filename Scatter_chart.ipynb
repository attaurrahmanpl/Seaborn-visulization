```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
data.shape
```

```python
data.head()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(data['Age'] , data['Fare'])
plt.show()
```

```python
# same chart to creat it more reabable
plt.figure(figsize=(16,9))
sns.scatterplot(y = data['Fare'] , x = data['Age'] , data = data , hue='Sex')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(y = data['Fare'] , x = data['Age'] , data = data , hue='Sex' , style='Embarked' , size='Embarked')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(y = data['Fare'] , x = data['Age'] , data = data , hue='Sex' , style='Embarked' , size='Embarked' , sizes=(100,400))
plt.show()
```
